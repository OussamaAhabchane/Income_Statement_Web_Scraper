```python
from bs4 import BeautifulSoup
import requests
from finstmt import IncomeStatements, BalanceSheets, FinancialStatements
import pandas as pd
import warnings
import matplotlib.pyplot as plt
```

```python
url = 'https://stockanalysis.com/stocks/aapl/financials/?p=quarterly'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
# Find all rows in the table
table_rows = soup.find_all('tr')

# Extract data from rows and store in DataFrame
data = []
columns = []

for row_index, row in enumerate(table_rows):
    if row_index == 0:
        header_cells = row.find_all('th')
        columns = [cell.get_text(strip=True) for cell in header_cells]
    else:
        cells = row.find_all(['th', 'td'])
        row_data = [cell.get_text(strip=True) for cell in cells]
        data.append(row_data)
        
columns[0] = ""

# Create a DataFrame
inc_df = pd.DataFrame(data, columns=columns)

inc_df = inc_df.iloc[:, :-1]


inc_df = inc_df.reset_index(drop=True)
inc_df
```

```python
# Assuming you have a DataFrame named 'df' that you want to save
excel_filename = 'inc_data.xlsx'  # Remplacez 'nom_du_fichier.xlsx' par le nom de fichier souhaité

# Save the DataFrame to an Excel file
inc_df.to_excel(excel_filename, index=False)  # Si vous souhaitez conserver l'index, changez index=False en index=True

print(f"DataFrame enregistré sous {excel_filename}")
```
